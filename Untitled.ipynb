```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
test_beneficiary_df = pd.read_csv('./HealthClaims/Test_Beneficiarydata-1542969243754.csv')
test_inpatient_df = pd.read_csv('./HealthClaims/Test_Inpatientdata-1542969243754.csv')
test_outpatient_df = pd.read_csv('./HealthClaims/Test_Outpatientdata-1542969243754.csv')
train_beneficiary_df = pd.read_csv('./HealthClaims/Train_Beneficiarydata-1542865627584.csv')
train_inpatient_df = pd.read_csv('./HealthClaims/Train_Inpatientdata-1542865627584.csv')
train_outpatient_df = pd.read_csv('./HealthClaims/Train_Outpatientdata-1542865627584.csv')
train_fraud_df = pd.read_csv('./HealthClaims/Train-1542865627584.csv')
```

```python
train_impatient_df_dropped=train_inpatient_df.drop(columns=columns_to_drop)
```

```python
impateint_outpatient_concatted_df=pd.concat([train_impatient_df_dropped,train_outpatient_df],axis=0)
```

```python
impateint_outpatient_merged_df=pd.merge(impateint_outpatient_concatted_df,train_fraud_df,on='Provider',how='left')
```

```python
impateint_outpatient_merged_df=pd.merge(train_beneficiary_df,impateint_outpatient_merged_df,on='BeneID',how='left')
```

```python
impateint_outpatient_merged_df['PotentialFraud']=impateint_outpatient_merged_df['PotentialFraud'].replace({'Yes':1,'No':0})
```

```python
train_inp_subset_df=train_inpatient_df[['BeneID', 'DiagnosisGroupCode', 'AdmissionDt', 'DischargeDt']]
```

```python
impateint_outpatient_merged_df1=pd.merge(impateint_outpatient_merged_df,train_inp_subset_df,on='BeneID',how='inner')
```

```python
for col in train_inpatient_df:
    if col not in impateint_outpatient_merged_df.columns:
        impateint_outpatient_merged_df[col]=train_inpatient_df[col]
```

```python
# Convert 'DOB' and 'ClaimStartDt' to datetime objects
impateint_outpatient_merged_df['DOB'] = pd.to_datetime(impateint_outpatient_merged_df['DOB'])
impateint_outpatient_merged_df['ClaimStartDt'] = pd.to_datetime(impateint_outpatient_merged_df['ClaimStartDt'])
impateint_outpatient_merged_df['ClaimEndDt'] = pd.to_datetime(impateint_outpatient_merged_df['ClaimEndDt'])

# Calculate ClaimSettleTime in days
impateint_outpatient_merged_df['ClaimSettleTime'] = (impateint_outpatient_merged_df['ClaimEndDt'] - impateint_outpatient_merged_df['ClaimStartDt']).dt.days

# Calculate age at the time of filing the claim
impateint_outpatient_merged_df['Age'] = ((impateint_outpatient_merged_df['ClaimStartDt'] - impateint_outpatient_merged_df['DOB']).dt.days / 365).astype(int)
```

```python
drop_cols = ['ClmDiagnosisCode_3','OperatingPhysician', 'OtherPhysician', 'ClmAdmitDiagnosisCode','BeneID', 'DOD', 'DOB','ClaimID', 'ClmDiagnosisCode_4', 'ClmDiagnosisCode_5', 'ClmDiagnosisCode_6', 'ClmDiagnosisCode_7', 'ClmDiagnosisCode_8','ClmDiagnosisCode_9', 'ClmDiagnosisCode_10', 'ClmProcedureCode_1', 'ClmProcedureCode_2', 'ClmProcedureCode_3', 'ClmProcedureCode_4', 'ClmProcedureCode_5', 'ClmProcedureCode_6', 'AdmissionDt', 'DischargeDt', 'DiagnosisGroupCode', 'ClaimStartDt', 'ClaimEndDt']
```

```python
impateint_outpatient_merged_df1=impateint_outpatient_merged_df.drop(drop_cols,axis=1)
# Replace values in the 'RenalDiseaseIndicator' column
impateint_outpatient_merged_df1['RenalDiseaseIndicator'] = impateint_outpatient_merged_df1['RenalDiseaseIndicator'].replace({'Y': 1, '0': 0})

# Convert the 'RenalDiseaseIndicator' column to integer
impateint_outpatient_merged_df1['RenalDiseaseIndicator'] = impateint_outpatient_merged_df1['RenalDiseaseIndicator'].astype(int)
```

```python
# Remove 'PRV' prefix from the 'Provider' column
impateint_outpatient_merged_df1['Provider'] = impateint_outpatient_merged_df1['Provider'].str.replace('PRV', '')

# Remove 'PHY' prefix from the 'AttendingPhysician' column
impateint_outpatient_merged_df1['AttendingPhysician'] = impateint_outpatient_merged_df1['AttendingPhysician'].str.replace('PHY', '')

# Convert the 'Provider' column to integer type
impateint_outpatient_merged_df1['Provider'] = impateint_outpatient_merged_df1['Provider'].astype(int)

# Convert the 'AttendingPhysician' column to integer type if appropriate
# impateint_outpatient_merged_df1['AttendingPhysician'] = impateint_outpatient_merged_df1['AttendingPhysician'].astype(int)
```

```python
from sklearn.impute import SimpleImputer 

object_columns=impateint_outpatient_merged_df1.select_dtypes(include='object').columns

print(object_columns)

imputer=SimpleImputer(strategy='most_frequent')

imputer.fit(impateint_outpatient_merged_df1[object_columns])

impateint_outpatient_merged_df1[object_columns]=imputer.transform(impateint_outpatient_merged_df1[object_columns])
```

```python
impateint_outpatient_merged_df1['DeductibleAmtPaid']=impateint_outpatient_merged_df1['DeductibleAmtPaid'].fillna(0.0)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Iterate over each column in the DataFrame
for column in impateint_outpatient_merged_df1.select_dtypes(include=['object']).columns:
    # Encode the column
    impateint_outpatient_merged_df1[column] = label_encoder.fit_transform(impateint_outpatient_merged_df1[column])
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(['InscClaimAmtReimbursed','ClaimSettleTime'],axis=1),df['InscClaimAmtReimbursed'],test_size=0.5,random_state=42)
```

```python
# Define the hyperparameter grids for each algorithm
param_grid = {
    # 'LinearRegression': {},
    'RandomForestRegressor': {'n_estimators': [50, 100, 200], 'max_depth': [None, 5, 10]},
    # 'AdaBoostRegressor': {'n_estimators': [50, 100, 200], 'learning_rate': [0.1, 0.5, 1.0]},
    # 'GradientBoostingRegressor': {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 7]},
    # 'XGBRegressor': {'n_estimators': [50, 100, 200], 'max_depth': [3, 5, 7], 'learning_rate': [0.01, 0.1, 0.5]},
    # 'KNeighborsRegressor': {'n_neighbors': [3, 5, 10], 'weights': ['uniform', 'distance']},
    # 'DecisionTreeRegressor': {'max_depth': [None, 5, 10]}
}

# Instantiate the estimator objects
estimators = {
    # 'LinearRegression': LinearRegression(),
    'RandomForestRegressor': RandomForestRegressor(),
    # 'AdaBoostRegressor': AdaBoostRegressor(),
    # 'GradientBoostingRegressor': GradientBoostingRegressor(),
    # 'XGBRegressor': XGBRegressor(),
    # 'KNeighborsRegressor': KNeighborsRegressor(),
    # 'DecisionTreeRegressor': DecisionTreeRegressor()
}

# Perform grid search for each algorithm
best_models = {}
for algorithm in param_grid.keys():
    grid_search = GridSearchCV(estimators[algorithm], param_grid[algorithm], cv=5, scoring='neg_mean_squared_error')
    grid_search.fit(X_train, y_train)
    best_models[algorithm] = grid_search.best_estimator_

# Evaluate the best modelsl
mse_results = {}
for algorithm, model in best_models.items():
    y_pred = model.predict(X_test)
    mse_results[algorithm] = mean_squared_error(y_test, y_pred)
```
