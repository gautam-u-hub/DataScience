```python
# Importing necessary libraries
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import xgboost as xgb
import pickle
import numpy as np

```

```python
# Loading data
beneficiary_data = pd.read_csv('./HealthClaims/Train_Beneficiarydata-1542865627584.csv')
inpatient_data = pd.read_csv('./HealthClaims/Train_Inpatientdata-1542865627584.csv')
outpatient_data = pd.read_csv('./HealthClaims/Train_Outpatientdata-1542865627584.csv')
fraud_data = pd.read_csv('./HealthClaims/Train-1542865627584.csv')

```

```python
# Dropping unnecessary columns from inpatient data
columns_to_drop = ['AdmissionDt', 'DischargeDt', 'DiagnosisGroupCode']
new_inpatient_data = inpatient_data.drop(columns=columns_to_drop)

# Combining inpatient and outpatient data
combined_data = pd.concat([new_inpatient_data, outpatient_data], axis=0)

# Merging combined data with fraud data
combined_data_with_fraud = pd.merge(combined_data, fraud_data, on='Provider', how='left')

```

```python
merged_data = pd.merge(beneficiary_data, combined_data_with_fraud, on='BeneID', how='left')

# Dropping columns with more than 50% missing values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data where ChronicCond_Diabetes is 1 or 2
diabetes_data = merged_data[merged_data['ChronicCond_Diabetes'].isin([1, 2])]

# Calculate average InscClaimAmtReimbursed for each category of ChronicCond_Diabetes
average_reimbursement = diabetes_data.groupby('ChronicCond_Diabetes')['InscClaimAmtReimbursed'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x='ChronicCond_Diabetes', y='InscClaimAmtReimbursed', data=average_reimbursement)
plt.xlabel('Chronic Diabetes Condition (1: Yes, 2: No)')
plt.ylabel('Average InscClaimAmtReimbursed')
plt.title('Average Reimbursement Amount for Diabetes Condition')
plt.show()
```

```python
cleaned_data = merged_data.dropna(thresh=len(merged_data) * 0.5, axis=1)
processed_data = cleaned_data

X = processed_data.drop(columns=['InscClaimAmtReimbursed', 'Provider'])
y = processed_data['InscClaimAmtReimbursed']

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, X.select_dtypes(include=['int', 'float']).columns),
        ('cat', categorical_transformer, X.select_dtypes(include=['object']).columns)
    ])

X_preprocessed = preprocessor.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

model = xgb.XGBRegressor(n_estimators=100, max_depth=5, learning_rate=0.1, random_state=42)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("MSE:", mse)
print("RMSE:", np.sqrt(mse))

r2 = r2_score(y_test, y_pred)
print("R^2 Score:", r2)
n = X_test.shape[0]  
p = X_test.shape[1]  
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
print("Adjusted R^2 Score:", adjusted_r2)

with open('./HealthClaims/healthcare_preprocessor.joblib', 'wb') as preprocessor_file:
    pickle.dump(preprocessor, preprocessor_file)

with open('./HealthClaims/xgboost_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)

model_filename = './HealthClaims/xgboost_model.pkl'
with open(model_filename, 'rb') as model_file:
    loaded_model = pickle.load(model_file)
```

```python
example_data = processed_data
first_row = example_data.iloc[[0]]  
X_first_row_preprocessed = preprocessor.transform(first_row)
predicted_value = loaded_model.predict(X_first_row_preprocessed)
actual_value = example_data.iloc[0]['InscClaimAmtReimbursed']
print("Predicted Value:", predicted_value)
print("Actual Value:", actual_value)

```

```python
# Saving modes of each column to a file
modes = example_data.mode()
modes_dict = {}
for col in modes.columns:
    mode_val = modes[col][0]  
    modes_dict[col] = [mode_val]  
modes_df = pd.DataFrame(modes_dict)
modes_df.to_csv('./HealthClaims/column_modes.csv', index=False)
print("Modes of each column have been saved to 'column_modes.csv' file.")

# Dropping columns with more than 50% missing values
print("Columns before dropping:")
print(merged_data.columns)
dropped_columns = merged_data.columns[merged_data.isnull().mean() > 0.5]
print("\nColumns dropped:")
print(dropped_columns)

```
